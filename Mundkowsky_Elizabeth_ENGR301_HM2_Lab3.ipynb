```python
from sklearn.model_selection import train_test_split
```

```python
x = p1[["AMBIENT_TEMPERATURE"]]
y = p1["DC_POWER"]
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

PART 2 QUESTION 1B

We created four arrays so that we can set aside 30% of the data (including the x and y values) for testing after running the model on the x and y training values to check if the model can accurately predict the y values from the x values. 

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

PART 2 QUESTION 1D

The R^2 value of my model is 0.529.

```python
plt.figure(figsize=(6,6))
plt.scatter(x_test,(y_test-y_pred))
plt.xlabel("Ambient Temperature (C)")
plt.ylabel("DC Power (kW)")
```

PART 2 QUESTION 1E

There is a large spread in the residual plot, and there appears to be a negative, linear trend, meaning the model predicts less accurately for higher temperatures and overestimates the DC power. This suggests the model is not valid.

```python
model.score(x_train,y_train)
```

PART 2 QUESTION 1F
Since the R^2 scores are similar between the test and training data, it seems I did avoid overfitting. This approach lets us know if I avoided overfitting because the model is about as accurate at predicting values on the test data as it is on the training data. If the model was overfitted, it would perform better on the training data.

```python
x1 = p1[["IRRADIATION"]]
y1 = p1["DC_POWER"]
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size = 0.3)
model1 = LinearRegression()
model1.fit(x_train1, y_train1)
y_pred1 = model1.predict(x_test1)
r2_score(y_test1,y_pred1)
```

PART 2 QUESTION 2A
The R^2 of this model is 0.978.

```python
plt.figure(figsize=(6,6))
plt.scatter(x_test1,(y_test1-y_pred1))
plt.xlabel("Irradiation (W/m^2)")
plt.ylabel("DC Power (kW)")
```

PART 2 QUESTION 2B
The spread of the residuals looks small and generally centered around zero for the entire range of irradiation, although there seems to be some outliers and greater spread around medium irradiation. This tells me the model is valid across the entire range of irradiation.

```python
model1.score(x_train1,y_train1)
```

PART 2 QUESTION 2C
Since the R^2 scores with the training and test data are similar, yes overfitting was avoided.

Some DC power sensors may need maintenance or replacement because they are producing zero readings when there is irradiation, which is most likely explained by faulty readings because the sensors resumed reading normally sporadically, indicating the actual solar panel is functional otherwise. At plant 1, 63 failure readings were recorded, while at plant 2, over 3000 failure readings were recorded, suggesting plant 2 needs more attention. From the non-null and non-zero sensor readings, it appears that DC power is most strongly correlated with irradiation and module temperature. Ambient temperature is also correlated with DC power output, but not as strongly. Daily yield indicates that the average peak cumulative power output is roughly 5000 kW. However, it is difficult to draw conclusions from the daily yield regarding the plants' function as the daily yield increases throughout the day and thus does not have a linear relationship with the sun conditions at any given point.
